```python
def normalize_text(s):
    #s = unicodeToAscii(s.lower().strip())
    #s=s.lower().strip()
    s = re.sub(r"([,.:!?])", r" \1 ", s)
    s = re.sub(r"[\|]+", r" ", s)
    s = re.sub("\n"," ", s)
#     s = re.sub('\n', '<NEWLINE>', s)
    s = re.sub('\t', ' ', s)
    #s = re.sub('\d+', '#', s)
    s = re.sub(' +', ' ',s)
    return s

def tokenizer_norm(text):
    text = tokenizer(normalize_text(text))
    return text
```
